```python
import csv
import os
path = './Final_Test/Images'

csv_reader = csv.reader(open(f"{path}/GT-final_test.csv"), delimiter = ';')
next(csv_reader)
for line in csv_reader:
    filename = line[0].replace('ppm','png')
    classid = int(line[7])
    if classid<10:
        class_f = f"0000{classid}"
    else:
        class_f =f"000{classid}"
    print(f'{path}/{class_f}/{filename}')
    os.rename(f'{path}/{filename}', f'{path}/{class_f}/{filename}')
```
